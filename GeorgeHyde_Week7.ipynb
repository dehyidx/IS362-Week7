### All user ratings

```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series

ratings = pd.read_csv('/home/gh23935521@sps.cuny.edu/Downloads/ratings.csv')
ratings
```

### Average ratings for each movie

```python
ratings.groupby('movie').rating.mean()
```

### Average ratings for each rater

```python
ratings.groupby('rater').rating.mean()
```

### All normalized rater ratings

```python
# copies the data 
norm_ratings = ratings.copy() 
  
# apply normalization formula  
norm_ratings['rating'] = (norm_ratings['rating'] - norm_ratings['rating'].min()) / (norm_ratings['rating'].max() - norm_ratings['rating'].min())
      
# view normalized data 
norm_ratings 
```

### Normalized movie averages

```python
norm_ratings.groupby('movie').rating.mean()
```

### Normalized rater averages

```python
norm_ratings.groupby('rater').rating.mean()
```

### Conclusion

Using normalized data has the advantage of placing measurements of different units and scale along the same scale for more accurate comparisons. However, it has the disadvantage of scaling the data to a smaller interval when outliers are present, therefore hindering it's accuracy. For example, when the compared to the original user averages, all the normal averages are more varied when compared to the original averages, where you can see that all the movies are well favored amongst the raters, as all of those averages are closer to 5.

### [Extra Credit] All standardized rater ratings

```python
std_ratings = ratings.copy()

# apply standardization formula  
std_ratings['rating'] = (std_ratings['rating'] - std_ratings['rating'].mean()) / (std_ratings['rating'].std())

std_ratings
```

### Standardized movie averages

```python
std_ratings.groupby('movie').rating.mean()
```

### Standardized User Averages

```python
std_ratings.groupby('rater').rating.mean()
```
